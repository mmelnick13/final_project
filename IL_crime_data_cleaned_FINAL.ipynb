```python
# Create function to determine what season the arrest took place
def get_season(date):
    if date > '2019-03-20' and date < '2019-06-20':
        return "spring"
    elif date > '2019-06-21' and date < '2019-09-21':
        return 'summer'
    elif date > '2019-09-21' and date < '2019-12-20':     
        return 'autumn'
    else:
        return "winter"
```

```python
# Create variable to return the season based on arrest date
season = arrestee_df["ARREST_DATE"].map(str).map(get_season)
```

```python
# Add a season column to arestee_df
arrestee_df.insert(loc= 5, column='SEASON', value = season)
arrestee_df.sample(n=5)
```

```python
# Write arestee_df DataFrame to a csv file.
arrestee_df.to_csv('IL/arrestee.csv', index=False)
```

```python
# Calculate number of crimes per season.
arrest_season = arrestee_df.groupby('SEASON')['ARRESTEE_SEQ_NUM'].agg(TOTAL_INCIDENT='sum')
arrest_season
```

```python
# Extract the season with the highest total of incidents.
high_crime_season = arrest_season[arrest_season.TOTAL_INCIDENT == arrest_season.TOTAL_INCIDENT.max()]
high_crime_season
```

```python
# Load the arrest_type.csv dataset.
arrest_types_df = pd.read_csv("IL/NIBRS_ARREST_TYPE.csv")
arrest_types_df
```

```python
# Load the offense_type dataset.
offense_types_df = pd.read_csv("IL/NIBRS_OFFENSE_TYPE.csv")
offense_types_df.head()
```

```python
# Load the offense.csv dataset.
offenses_df = pd.read_csv("IL/NIBRS_OFFENSE.csv")
offenses_df.head()
```

```python
# Load the location_type.csv dataset.
location_df = pd.read_csv("IL/NIBRS_LOCATION_TYPE.csv")
location_df.drop('LOCATION_CODE', axis=1, inplace=True)
location_df.head()
```

```python
# Join arestee_df and arrest_types_df on ARREST TYPE ID
joined_df = arrestee_df.merge(arrest_types_df, on ="ARREST_TYPE_ID")
```

```python
# Join joined_df and offense_types on OFFENSE_TYPE_ID.
joined_df2 = joined_df.merge(offense_types_df, on ="OFFENSE_TYPE_ID")
```

```python
# Merge joined_df2 and offenses_df on INCIDENT_ID.
joined_df3 = joined_df2.merge(offenses_df, on ="INCIDENT_ID")
```

```python
# Merge joined_df3 and location_df on LOCATION_ID.
joined_df4 = joined_df3.merge(location_df, on ="LOCATION_ID")
```

```python
# Drop duplicate columns
IL_crime_joined = joined_df4.drop(['DATA_YEAR_x','OFFENSE_TYPE_ID_x','DATA_YEAR_y','OFFENSE_TYPE_ID_y'], axis = 1)
```
